```python
# Import necessary libraries

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt





# Load the dataset

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')



# Display basic information about the dataset

print(df.info())





# # Check the number of NaN values in all columns







# Check the number of NaN values in 'TotalCharges'

num_nan = df['TotalCharges'].isnull().sum()

print(f"Number of NaN values in 'TotalCharges': {num_nan}")




























```

```python
#Even though we dont find any null values. there can be blank values



#blank_entries_per_column = (df == '').sum()



blank_entries_per_column = (df == ' ').sum()

print(blank_entries_per_column)




```

```python
# The 11 non numeric values will be changed to null values.

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')



# Step 2: Fill missing values in 'TotalCharges' with the median value

df.dropna(inplace=True)
#df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)



#setting all columnnames to lowercase.
df.columns = map(str.lower, df.columns)

df.head()

df.drop(['customerid'], axis=1, inplace=True)
df = df.drop(['gender'], axis=1)
#df.drop(['huge_family'], axis=1, inplace=True)
#df.drop(['streaming'], axis=1, inplace=True)

df.drop(['streamingtv'], axis=1, inplace=True)
df = df.drop(['phoneservice'], axis=1)
df = df.drop(['onlinesecurity'], axis=1)
```

```python
df.head()
```

```python
# Step 3: Convert categorical columns to numerical values using one-hot encoding

# categorical_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',

#                        'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',

#                        'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',

#                        'PaperlessBilling', 'PaymentMethod', 'Churn']

categorical_columns = ['partner', 'dependents', 'multiplelines',

                       'internetservice', 'onlinebackup', 'deviceprotection',

                       'techsupport', 'streamingmovies', 'contract',

                       'paperlessbilling', 'paymentmethod']



# Apply one-hot encoding to these columns

df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
print(df.head())
```

```python


# Step 4: Standardize 'tenure', 'MonthlyCharges', and 'TotalCharges' (optional but useful)

from sklearn.preprocessing import StandardScaler



scaler = StandardScaler()

df[['tenure', 'monthlycharges', 'totalcharges']] = scaler.fit_transform(df[['tenure', 'monthlycharges', 'totalcharges']])



# Step 5: Drop unnecessary columns

#df.drop('customerID', axis=1, inplace=True)



# Display the first few rows of the preprocessed dataset

print(df.head())
```

```python
print(df.info())
```

```python
from sklearn.model_selection import train_test_split



# Features (X) and Target (y)
df['churn'] = df['churn'].map({'Yes': 1, 'No': 0})
X = df.drop(columns=['churn'])
y = df['churn']



# Train-test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
```

```python
y.head(1)
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score, classification_report



# Random Forest model

rf = RandomForestClassifier(random_state=42)



# Hyperparameter grid

param_grid = {

    'n_estimators': [100, 200, 300],

    'max_depth': [10, 20, 30],

    'min_samples_split': [2, 5, 10],

}



# GridSearchCV

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,

                           cv=5, scoring='accuracy', verbose=2, n_jobs=-1)



# Fit on training data

grid_search.fit(X_train, y_train)



# Best parameters

print("Best Parameters:", grid_search.best_params_)



```

```python
# Best estimator
best_params = grid_search.best_params_
#best_rf = grid_search.best_estimator_
rf_classifier_best = RandomForestClassifier(**best_params, random_state=42)
rf_classifier_best.fit(X_train, y_train)
```

```python
train_accuracy = rf_classifier_best.score(X_train, y_train)
print(train_accuracy)
test_accuracy = rf_classifier_best.score(X_test, y_test)
print(test_accuracy)
```

```python
#
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define classifiers and their hyperparameter grids
classifiers_and_params = {
    'RandomForest': {
        'model': RandomForestClassifier(random_state=42),
        'param_grid': {
            'n_estimators': [100, 200, 300],
            'max_depth': [10, 20, 30],
            'min_samples_split': [2, 5, 10],
        }
    },
    'KNN': {
        'model': KNeighborsClassifier(),
        'param_grid': {
            'n_neighbors': [3, 5, 7],
            'weights': ['uniform', 'distance'],
            'metric': ['euclidean', 'manhattan']
        }
    },
    'SVM': {
        'model': SVC(random_state=42),
        'param_grid': {
            'C': [0.1, 1, 10],
            'kernel': ['linear', 'rbf', 'poly'],
            'gamma': ['scale', 'auto']
        }
    },
    'GradientBoosting': {
        'model': GradientBoostingClassifier(random_state=42),
        'param_grid': {
            'n_estimators': [100, 200],
            'learning_rate': [0.05, 0.1, 0.2],
            'max_depth': [3, 5, 7]
        }
    }
}

# Loop through classifiers and tune parameters
for name, config in classifiers_and_params.items():
    print(f"Running GridSearchCV for {name}...")
    
    model = config['model']
    param_grid = config['param_grid']
    
    # Grid search
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, 
                               cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    
    # Best parameters and model
    best_params = grid_search.best_params_
    best_model = grid_search.best_estimator_
    
    # Fit best model on training data
    best_model.fit(X_train, y_train)



    print(f"Best Parameters for {name}: {best_params}")
    y_test_pred = best_model.predict(X_test)

    # Calculate accuracy
    #train_accuracy = best_model.score(X_train, y_train)
    #test_accuracy = best_model.score(X_test, y_test)

    accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred)
    recall = recall_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred)
    
    
    # Print metrics
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1-Score: {f1:.2f}")
    print("-" * 50)
```
